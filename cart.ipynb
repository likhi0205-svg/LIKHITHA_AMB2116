```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("iris.csv")
df
```

```python
df["Species"].unique() #data frame   gives the unique values in the coloum
```

```python
 df["Species"].value_counts() #gives the number of counts of each species
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,4].values
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train ,x_test,y_train ,y_test =train_test_split(x,y,test_size = 1/3,random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
y_test
```

```python
y_train
```

```python

from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(criterion="entopy",max_depth=None,min_samples_split=5,min_samples_leaf=2)


```

```python
#train model
model.fit(x_train,y_train)
print(model)
```

```python
#predict the testing model

y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
c = confusion_matrix(y_test,y_pred)
c
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
      
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
#visualization

from sklearn import tree
fn = ["sepal length(cm)","sepal width(cm)","petal length(cm)","petal width(cm)"]
cn = ["setosa","versicolor","virginica"]
fig, ages = plt.subplots(nrows = 1,ncols=1,figsize = (4,4), dpi =300)
tree.plot_tree(model,
               feature_names = fn,
               class_names = cn,
               filled = True);
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
dataset = pd.read_csv("Mall_Customers.csv")
dataset
```

```python



X=dataset.iloc[:,[3,4]].values
dataset






```

```python
X=dataset.iloc[:,[3,4]].values
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=6,init='k-means++',random_state=42)
y_kmeans=kmeans.fit_predict(X)
y_kmeans

```

```python
len(y_kmeans)
```

```python
print(np.round(kmeans.cluster_centers_[:0],3))
print(np.round(kmeans.cluster_centers_[:,1],3))

```

```python
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='yellow',label='Centroids')
plt.legend()
plt.title("Clusters of customers")
plt.xlabel("Annual income(k$")
plt.ylabel("spending score(1-100)")
plt.show()
```

```python
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='yellow',label='Centroids')
plt.legend()
plt.title('Clusters of customers')
plt.xlabel('Annual Income (K$)')
plt.ylabel('Spending score(1-100)')
plt.show()
```

```python
plt.scatter(X[y_kmeans==0,0],X[y_kmeans==0,1],c='red',label='Cluster 1')
plt.scatter(X[y_kmeans==1,0],X[y_kmeans==1,1],c='green',label='Cluster 2')
plt.scatter(X[y_kmeans==2,0],X[y_kmeans==2,1],c='blue',label='Cluster 3')
plt.scatter(X[y_kmeans==3,0],X[y_kmeans==3,1],c='cyan',label='Cluster 4')
plt.scatter(X[y_kmeans==4,0],X[y_kmeans==4,1],c='magenta',label='Cluster 5')
plt.scatter(X[y_kmeans==5,0],X[y_kmeans==5,1],c='pink',label='Cluster 6')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='yellow',label='Centroids')
plt.legend()
plt.title('Clusters of customers')
plt.xlabel('Annual Income (K$)')
plt.ylabel('Spending score(1-100)')
plt.show()
```

```python
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='yellow',label='Centroids')
plt.legend()
plt.title('Clusters of customers')
plt.xlabel('Annual Income (K$)')
plt.ylabel('Spending score(1-100)')
plt.show()
```

```python
#Elbow method to find the optimal number of clusters
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=1)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,11),wcss)
plt.title('The Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
    
print(wcss)
```
