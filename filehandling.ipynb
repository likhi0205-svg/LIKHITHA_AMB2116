```python
def expr(a,b):
    try:
        c=a/b
        print(c)
    except ZeroDivisionError:
        print("divided by 0")
    except NameError:
        print("variable not declared")
    except:
        print("some error occured")
    else:
        print("excecuted without parameter")

    finally:print("end of the program")

a=3
b=5
expr(a,b)
```
