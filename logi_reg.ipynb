```python
#slicing the data set
x = df.iloc[:,[2,3]].values
y = df.iloc[:,4].values
print(y)
print(np.shape(x))
print(df.head())
```

```python
#spliting the gataset into training set and testing set
from sklearn.model_selection import train_test_split
x_train ,x_test,y_train,y_test = train_test_split(x,y,test_size = 1/3,random_state=0)
```

```python
print(x_train[0:10])
from sklearn.preprocessing import StandardScaler
stdsc = StandardScaler()
x_train_std = stdsc.fit_transform(x_train)
x_test_std = stdsc.transform(x_test)
print(x_train_std[0:10])
```

```python
x_test.size
```

```python
y_test.size
```

```python
x_train.size
```

```python
y_train.size
```

```python
#create an object of the algorithm/model
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=0)
print(clf)
```

```python
#Train the model
clf.fit(x_train_std,y_train)
```

```python
#predict the model on testing dataset
y_pred = clf.predict(x_test_std)
print(y_pred)
y_pred
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy:",accuracy_score(y_test,y_pred))
```

```python
# Visualising the Training set results
from matplotlib.colors import ListedColormap
X_set, y_set = x_train_std, y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, clf.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
