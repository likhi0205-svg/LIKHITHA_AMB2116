```python
import numpy as np
crop_yield=np.array([
    [2,3,4,5],
    [1,12,34,56],
    [32,43,54,65],
    [22,33,44,76]
   ])


print(crop_yield)
print(f'\n data type:{crop_yield.dtype}')
print(f'shape:{crop_yield.shape}')
print(f'total elements:{crop_yield.size}')
print(f'number of dimension:{crop_yield.ndim}')

    
```

```python
(2+3+4+5)/4
```

```python
total_yield=np.sum(crop_yield)
avg_yield_per_farm = np.mean(crop_yield,axis=1)
avg_yield_per_season= np.mean(crop_yield,axis=0)
max_yield = np.max(crop_yield)
min_yield = np.min(crop_yield)
print("total production:",total_yield)
print("avg yield axis =1:",avg_yield_per_farm)
print("avg yield axis =0:",avg_yield_per_season)
print("maximum yield:",max_yield)
print("minimum yield:",min_yield)
```

```python
max_yield = np.max(crop_yield,axis=1)
min_yield = np.min(crop_yield,axis=0)
print("maximum yield:",max_yield)
print("minimum yield:",min_yield)
```

```python
max_yield = np.max(crop_yield,axis=0)
min_yield = np.min(crop_yield,axis=1)
print("maximum yield:",max_yield)
print("minimum yield:",min_yield)
```

```python
best_farm=np.argmax(avg_yield_per_farm)
print("\nbest performing farm:farm",best_farm+1)
print("thier average yield:",round(avg_yield_per_farm[best_farm],2),"tons")
```

```python
#boolen indexing:find exceptional yields
import numpy as np
crop_yield=np.array([
    [2,3,4,5],
    [1,12,34,56],
    [32,43,54,65],
    [22,33,44,76]
   ])
print(crop_yield)
exceptional =crop_yield>34
print(exceptional)
print("\n exceptional  yields(>34):")
print(crop_yield[exceptional])
```
