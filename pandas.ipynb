```python
import numpy as np
crop_yield=np.array([
    [2,3,4,5],
    [1,12,34,56],
    [32,43,54,65],
    [22,33,44,76]
   ])


print(crop_yield)
print(f'\n data type:{crop_yield.dtype}')
print(f'shape:{crop_yield.shape}')
print(f'total elements:{crop_yield.size}')
print(f'number of dimension:{crop_yield.ndim}')
```

```python
exceptional =crop_yield>34
print(exceptional)
print("\n exceptional  yields(>34):")
print(crop_yield[exceptional])
```

```python
#conditional operations:classify yields
categories =np.where(crop_yield>34,'high',
                     np.where(crop_yield>20,'mediun','low'))
print("\nyield categories:")
print(categories)
                     
```

```python
cumulative_yield = np.cumsum(crop_yield)
print(cumulative_yield)
```

```python
import pandas as pd
import numpy as np

#creating data with missing values
data = pd.DataFrame({
    'plot' : ['a','b','c','d','e'],
    'yield_kg' : [2500,np.nan,2599,np.nan,2600], #nan is not a number
    'rainfall' : [150,200,np.nan,180,170],
    'tempr' : [28,30,29,np.nan,24]
})

print("===data with missing values==")
print(data)

#detecting missing data
print("\ndetecting missing values")
print("ts null?")
print(data.isnull()) #sshowing wheather space is null or filled
print(f"\n total missing per column:")
print(data.isnull().sum()) #counting total nan
                    
```

```python
print("==filling strategies===")
data_mean = data.copy()
print(data_mean['yield_kg'])
data_mean['yield_kg'] = data_mean['yield_kg'].fillna(data_mean['yield_kg'].mean()) #fill with the number in null space
print("\n fill with mean:")
print(data_mean['yield_kg'])
```

```python
print("==filling strategies===") #fill the null space using median
data_median= data.copy()
print(data_median['rainfall'])
data_median['rainfall'] = data_median['rainfall'].fillna(data_median['rainfall'].median()) #fill with the number in null space
print("\n fill with median:")
print(data_median['rainfall'])
```

```python
#forward fill (use previous value)
data_ffill = data.copy()
print(data_ffill)
data_ffill = data_ffill.fillna(method='ffill')
print("\n forward fill:")
print(data_ffill)

#ffill -it replace missing values(nan)
#with the lat valid (non -missing )value above it in the column
```

```python
print("==droping the missing rows===")
data_dropped = data.dropna()
print(data_dropped)
print("original rows :",len(data))
print("after drop:",len(data_dropped))
```
