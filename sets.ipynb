#add element to set using update()

*once a set is created we can't change item ,but we can add new items.

```python
s1={'e','r','r','t','w','d'}
s2={'g','j','l','n','x','p'}
print("s1-->",s1)
print("s2-->",s2)
s1.update(s2)
print("after update-->",s1)
```

```python
s={'e','r','r','t','w','d'}
print("s  -->",s)
s.update(['l','n','p'])
print("after updates-->",s)
```

#Operations on sets
union,intersection,difference,symmetric difference

```python
s={'e','r','t','w','d'}
p={'l','n','j','k','x'}
print("s-->",s)
print("p-->",p)
print("union of s and p -->",s.union(p))
print("intersection of s and p -->",s.intersection(p))
print("difference s-p -->",s-p)
print("difference p-s -->",p-s)
print("symmetric difference s-p  -->",s.symmetric_difference(p))
print("symmetric difference p-s  -->",p.symmetric_difference(s))
```
