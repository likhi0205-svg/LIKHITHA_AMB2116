```python
str ="Today is 19-Jan-2026 monday"
b=v=c=d=o=0
for i in str:
    if i==' ':
        b=b+1

    elif i in 'aeiou':
        v=v+1

    elif i.isalpha():
        c=c+1

    elif i.isdigit():
        d=d+1

    else:
        o=o+1
print('number of blanks',b)     
print('number of vowels',v)
print("number of consonants",c)
print("number of digits",d)
print("others",o)
```

#Extracting all strings with prefix

```python
codes=[
'CRP_RICE_2026',
'CRP_WHEAT_2025',
'FRT_MANGO_2006',
'CRP_MAIZE_2024'
]
filtered=[]
for crop_code in codes:
    if crop_code.startswith("CRP"):
        filtered.append(crop_code)
print(filtered)
```

```python
codes=[
'CRP_RICE_2026',
'CRP_WHEAT_2025',
'FRT_MANGO_2006',
'CRP_MAIZE_2024'
]
clean_codes=[]
for code in codes:
    if code.startswith("CRP"):
        clean_codes.append(code[4:])
print(clean_codes)
```

```python
codes=[
'CRP_RICE_2026',
'CRP_WHEAT_2025',
'FRT_MANGO_2006',
'CRP_MAIZE_2024'
]
for crop_code in codes:
    if crop_code.startswith("CRP_"):
        code,crop,year=crop_code.split("_")
        print(crop,year)
        
```
