TUPLE WITH ONE ITEM

```python
T =('likhitha',) #comma should be used
print("T :",T)
print("totl number of elements :", len(T))
print("type of object :", type(T))
```

```python
T =('likhitha') #not used  comma lead to string data type
print("T :",T)
print("totl number of elements :", len(T))
print("type of object :", type(T))
```

TUPLE ITERATION USING WHILE LOOP

```python
T = tuple(['likhi','megha','lekhana',17,18]) #coversion of list to tuopl
print("T :",T)
i = 0
print(" ")
while i < len(T):
    print(T[i])
    i+=1
```

```python
T = tuple(['likhi','megha','lekhana',17,18]) #for loop
print("T :",T)
for i in range(len(T)):
    print(T[i])
    
```

```python
T = tuple(['likhi','megha','lekhana',17,18]) #coversion of list to tuple
print("T :",T)
for s in T:
    print(s)
```
